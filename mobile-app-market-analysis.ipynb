### Removing Rows with Missing Columns

The Google Play dataset has a dedicated [discussion section](https://www.kaggle.com/datasets/lava18/google-play-store-apps/discussion?sort=undefined), and [one of the discussions](https://www.kaggle.com/datasets/lava18/google-play-store-apps/discussion/66015) outlines an error for row 10472. The row is printed below and compared against the header.

#### Option 1 - .index() function

Python's built-in index() function is a useful tool for finding the index of a specific element in a sequence. This function takes an argument representing the value to search for and returns the index of the **first occurrence** of that value in the sequence.

```python
for row in android:
    if len(row) != len(android_header):
        index = android.index(row)
        print(f"Index: {index}") # f-string, which is a clean way to insert variables directly into a text string.
        print(f"Row Data: {row}")
        print('\n')
```

Using .index() inside a loop can be highly inefficient for large datasets. Because .index() searches through the dataset from the beginning each time to find the match, it significantly increases the computational time.

Also, the .index() function always returns the index of the **very first match** it finds. If the dataset happens to have duplicate rows, and row 10472 is identical to row 500, .index() will return 500, even though the loop is currently looking at row 10472!

#### Option 2 - enumerate() function

A much more efficient approach is to use Python's built-in enumerate() function. This provides both the index and the row data simultaneously as you iterate, completely bypassing the need to scan the list for the index separately.

```python
for index, row in enumerate(android):
    if len(row) != len(android_header):
        print(f"Row {index} has missing columns")
        print(f"Correct total columns: {len(android_header)}")
        print(f"Total columns in row {index}: {len(row)}")
        print('\n')
        print(f"header: {android_header}")
        print('\n')
        print(f"Row {index}: {row}")
        
```

### Removing Duplicate Entries

The Google Play dataset contains apps with multiple entries. These duplicates fall into two categories:

- Updated data: apps like Instagram have multiple entries because they were scraped at different times, resulting in varying review counts
- Identical copies: apps like Box have multiple entries that are exactly the same across every column

```python
print(android_header)

print('\n')

for app in android:
    name = app[0]
    if name == "Instagram":
        print(app)

print('\n')

for app in android:
    name = app[0]
    if name == "Box":
        print(app)
```

```python
# Identify duplicate apps in the Google Play dataset

duplicate_google_apps = []
unique_google_apps = []

for app in android:
    name = app[0]
    if name in unique_google_apps:
        duplicate_google_apps.append(name)
    else:
        unique_google_apps.append(name)

print("Number of duplicate apps:", len(duplicate_google_apps))
print('\n')
print("Example of duplicate apps:", duplicate_google_apps[:10])
```

The Google Play dataset contains 1,181 duplicate entries. Rather than removing duplicates randomly, the row with the highest number of reviews is kept for each app. The higher the review count, the more recent and reliable the data.

The approach taken:

- Build a dictionary where each key is a unique app name and the value is the highest review count found for that app
- Use the dictionary to create a cleaned dataset containing only one entry per app, keeping the row with the highest review count


```python
reviews_max = {}

for app in android:
    name = app[0]
    n_reviews = float(app[3])

    if name in reviews_max and reviews_max[name] < n_reviews:
        reviews_max[name] = n_reviews
    
    elif name not in reviews_max:
        reviews_max[name] = n_reviews
        
```

There are 1,181 cases where an app occurs more than once, so the length of the dictionary of unique apps should equal the difference between the total dataset length and 1,181.

```python
print("Expected row:", len(android)-1181)
print(len(reviews_max))
```

The `reviews_max` dictionary is used to remove duplicates, keeping only the entry with the highest review count for each app.

The logic:

- Two empty lists are initialised: `android_clean` and `already_added`
- For each app in the dataset, the name and review count are extracted
- The row is added to `android_clean` only if its review count matches the maximum recorded in `reviews_max` and the app name has not already been added
- The second condition handles cases where a duplicate app has identical review counts across all entries, such as Box, which would otherwise still produce duplicates

```python
android_clean = []
already_added = []

for app in android:
    name = app[0]
    n_reviews = float(app[3])

    if (reviews_max[name] == n_reviews) and (name not in already_added):
        android_clean.append(app)
        already_added.append(name) # make sure this is inside the if block
```

A quick check confirms the cleaned dataset contains 9,659 rows, as expected.

```python
explore_data(android_clean,0,3,True)
```

```python
# Identify duplicate apps in the Apple Store dataset

duplicate_ios_apps = []
unique_ios_apps = []

for app in ios:
    name = app[0]
    if name in unique_ios_apps:
        duplicate_ios_apps.append(name)
    else:
        unique_ios_apps.append(name)

print("Number of duplicate apps:", len(duplicate_ios_apps))
print('\n')
print("Example of duplicate apps:", duplicate_ios_apps[:10])
```

### Removing Non-English Apps

Several non-English apps are present in both datasets. Since this analysis focuses on the English-speaking market, these entries are removed.

```python
# Some examples of noon-English apps
print(ios[813][1])
print(ios[6731][1])

print('\n')

print(android_clean[4412][0])
print(android_clean[7940][0])
```

One way to identify non-English apps is to use Python's [built-in ord() function](https://docs.python.org/3/library/functions.html#ord). According to the [ASCII system](https://en.wikipedia.org/wiki/ASCII), common English characters fall within the range of 0 to 127. A function can be built to check each character in an app name against this range. Any character with a value above 127 is considered non-English.

**For context**, English text usually includes letters from the English alphabet, numbers composed of digits from 0 to 9, punctuation marks (., !, ?, ;, etc.), and other symbols (+, *, /, etc.).

```python
# Create a function to determine if a string is English. 

def is_english(string):
    for character in string:
        if ord(character) > 127:
            return False
            
    # 'return True' must sit here at the for-else level, not inside the if block.
    # Placing it inside would cause the function to return True on the very first
    # English character, exiting before any non-English character is ever checked.
    # e.g. "Instagr汉m" would incorrectly return True on the first letter "I".
    else: 
        return True
```

```python
print(is_english('Docs To Go™ Free Office Suite'))
print(is_english('Instachat 😜'))
```

The function works, but some legitimate English app names use emojis or special symbols (™, em dash, en dash, etc.) that fall outside the ASCII range. In its current form, the function would incorrectly remove these apps.

To minimise data loss, an app is only removed if its name contains **more than three** non-ASCII characters:


```python
def is_english(string):
    non_ascii = 0
    
    for character in string:
        if ord(character) > 127:
            non_ascii += 1
    
    if non_ascii > 3:
        return False
    else:
        return True

print(is_english('Docs To Go™ Free Office Suite'))
print(is_english('Instachat 😜'))
        
```

The filter is not perfect. A small number of non-English apps may still pass through, but this level of precision is sufficient for the purposes of this analysis.

The `is_english()` function is applied below to filter both datasets:

```python
android_english = []
ios_english = []

for app in android_clean:
    name = app[0]
    if is_english(name):
        android_english.append(app)

for app in ios:
    name = app[1]
    if is_english(name):
        ios_english.append(app)


explore_data(android_english, 0, 3, True)
print('\n')
explore_data(ios_english, 0, 3, True)
```

After filtering, 9,614 Android apps and 6,183 iOS apps remain.

### Most Common Apps by Genre

The analysis starts by identifying the most common genres in each market. A frequency table is built for the `prime_genre` column of the App Store dataset, and the `Genres` and `Category` columns of the Google Play dataset.

```python
# ================================================
# FUNCTION 1: Frequency Generator
# Calculates the percentage of apps in each genre
# ================================================
def freq_table(dataset, index):
    table = {} # Empty dictionary to store the raw frequency counts
    total = 0  # Accumulator to track the total number of rows

    # 1. Aggregate raw frequencies
    for row in dataset:
        total += 1             # Increment the total for each row
        value = row[index] # Extract the target variable (e.g., genre)
        # Populate the dictionary with counts
        if value in table:
            table[value] += 1
        else: 
            table[value] = 1

    # 2. Compute percentages
    table_percentages = {}
    for key in table:
        # Calculate the relative frequency as a percentage
        percentage = (table[key]/total) *100
        table_percentages[key] = percentage

    return table_percentages

# =========================================================
# FUNCTION 2: Display Output
# Sorts and prints the frequency table in descending order
# =========================================================
def display_table(dataset, index):
    # Retrieve the computed percentages
    table = freq_table(dataset, index)

    # Convert the dictionary into a list of tuples for sorting
    table_display = []
    for key in table:
        # Structure as (percentage, category_name) to sort numerically
        key_val_as_tuple = (table[key], key) #Tuple here
        table_display.append(key_val_as_tuple)

    # Sort the list of tuples in descending order
    table_sorted = sorted(table_display, reverse = True)

    # Print the sorted distribution
    for entries in table_sorted:
        print(entries[1] , ":", entries[0])
```

**Why use a tuple instead of a dictionary or a list?**

**1. Why not just sort the dictionary?**
Dictionaries are built for speed and key-based lookups, not for maintaining order. Passing a dictionary directly to Python's sorted() function will result in Python ignoring the values (the percentages) and only sorting the keys (the genre names) in alphabetical order. Because the objective is to sort the data numerically by percentage, the dictionary format is insufficient.

**2. Why a tuple instead of a list?** Technically, structuring the data as a list of lists ([percentage, genre]) instead of a list of tuples ((percentage, genre)) would still work, as Python sorts both data types by evaluating the first item. However, data scientists use tuples in this scenario to communicate **intention**.

Lists are mutable; they are designed for dynamic data that may be appended, removed, or reorganised later. Tuples are immutable and are meant for fixed records. Because a specific genre is permanently locked to its specific percentage in this dataset, a tuple is the safest and most "Pythonic" way to package them together. It clearly signals that this pair is a finalised record that should not be altered.

### Most Popular Apps by Genre on the App Store



One way to identify the most popular genres is to calculate the **average number of installs** per genre. The Google Play dataset includes an `Installs` column, but this information is absent from the App Store dataset. The total number of user ratings (`rating_count_tot`) is used as a proxy.

#### Apple Store Dataset


```python
# Get all unique genres from the iOS dataset (column -5 = genre column)
genre_ios = freq_table(ios_final, -5)

# Empty dictionary to store results
# Need this dictionary ONLY needed for sorting at the end, not during calculation

# Step 1: Calculate average ratings per genre
avg_ratings_by_genre = {}

for genre in genre_ios:
    total = 0       # Reset running total for this genre
    len_genre = 0   # Reset app counter for this genre
    for app in ios_final:
        genre_app = app[-5]
        if genre_app == genre: # Check if this app belongs to the genre currently being checked
            n_rating = float(app[5])
            total += n_rating
            len_genre += 1
    avg_n_rating = total / len_genre
    avg_ratings_by_genre[genre] = avg_n_rating # Store result in dictionary e.g. {'Games': 22788.67, 'Music': 57326.53}
    
# Sort by value (highest to lowest) and print
# .items()           → returns (genre, number) pairs from the dictionary
# key=lambda x: x[1] → sort by number (position 1), not genre name (position 0)
# reverse=True       → highest first
for genre, avg_n_rating in sorted(avg_ratings_by_genre.items(), key = lambda x: x[1], reverse = True):
        print(genre, ":", avg_n_rating)
```

On average, navigation apps have the highest number of user ratings, but this figure is heavily skewed by Waze and Google Maps, which together account for close to half a million reviews.

```python
for app in ios_final: 
    if app[-5] == "Navigation":
        print(app[1], ":", app[5]) # Print name and number of ratings
```

The same pattern applies to social networking apps, where the average is skewed by giants like Facebook, Pinterest, and Skype, and to music apps, where Pandora, Spotify, and Shazam dominate.

These genres may appear more popular than they actually are. The averages are pulled up by a small number of apps with hundreds of thousands of ratings, while most other apps in the same genre struggle to reach 10,000. A more accurate picture could be obtained by removing these outliers before recalculating, which is left for a future iteration.

Reference apps have 74,942 user ratings on average, but it's actually the Bible and Dictionary.com which skew up the average rating:

```python
for app in ios_final: 
    if app[-5] == "Reference":
        print(app[1], ":", app[5]) # Print name and number of ratings
```

Despite the skew, this genre shows potential. One opportunity is to build an app centred around a popular book, adding features beyond the raw text such as daily quotes, an audio version, chapter quizzes, or an embedded dictionary. #### Google Play Dataset


The Google Play dataset includes install data, which gives a more direct measure of genre popularity. However, the values are open-ended (100+, 1,000+, 5,000+, etc.) and are not precise:

```python
display_table(android_clean, 5)
```

The install numbers are open-ended (100+, 1,000+, 5,000+, etc.), so exact figures are unavailable. For this analysis, each value is treated as its lower bound. An app with 100,000+ installs is counted as 100,000, and an app with 1,000,000+ installs is counted as 1,000,000.

Before computation, the comma and plus characters are stripped from each value and the result is converted to float. This is handled directly in the loop below, where the average number of installs per genre is also calculated.

```python
genre_android = freq_table(android_final, 1)

avg_n_installs_by_genre = {}

for genre in genre_android:
    total = 0
    len_genre = 0
    for app in android_final:
        genre_app = app[1]
        if genre_app == genre:
            n_install = app[5]
            n_install = n_install.replace(",","")
            n_install = n_install.replace("+","")
            total += float(n_install)
            len_genre += 1
    
    avg_n_installs = total / len_genre
    avg_n_installs_by_genre[genre] = avg_n_installs

for genre, avg_n_installs in sorted(avg_n_installs_by_genre.items(), key = lambda x: x[1], reverse = True):
    print(genre, ":", avg_n_installs)
```
